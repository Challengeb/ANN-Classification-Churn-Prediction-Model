```python
import tensorflow as tf
from tensorflow.keras.models import load_model
import pickle
import pandas as pd
import numpy as np
```

```python
## load the trained model, scaler pickle, onehot
model = load_model("model.h5")
```

```python
## load the encoder and the scaler 
with open("onehot_encoder_geo.pkl", "rb") as file:
    label_encoder_geo = pickle.load(file)

with open("label_encoder.pkl", "rb") as file:
    label_encoder_gender = pickle.load(file)


with open("scaler.pkl", "rb") as file:
    scaler = pickle.load(file)
```

```python
input_data = {
    "CreditScore": 600,
    "Geography": "France",
    "Gender": "Male",
    "Age": 40,
    "Tenure": 3,
    "Balance": 60000,
    "NumOfProducts": 2,
    "HasCrCard": 1,
    "IsActiveMember": 1, 
    "EstimatedSalary": 50000
}
```

```python
# One-hot encode "Geography"
geo_encoded = label_encoder_geo.transform([[input_data["Geography"]]]).toarray()
geo_encoded_df = pd.DataFrame(geo_encoded, columns=label_encoder_geo.get_feature_names_out(["Geography"]))
```

```python
# One-hot encode "Geography"
# Remove .toarray() because the output is already a dense array
geo_encoded = label_encoder_geo.transform([[input_data["Geography"]]])

# Now create the DataFrame
geo_encoded_df = pd.DataFrame(
    geo_encoded, 
    columns=label_encoder_geo.get_feature_names_out(["Geography"])
)
```

```python
geo_encoded_df
```

```python
input_df = pd.DataFrame([input_data])
input_df
```

```python
# Encode my categorical variables
input_df["Gender"]=label_encoder_gender.transform(input_df["Gender"])
input_df
```

```python
# concatination one hot encoded data
input_df = pd.concat([input_df.drop("Geography", axis=1), geo_encoded_df], axis=1)
```

```python
input_df
```

```python
## Scaling the input data
input_scaled = scaler.transform(input_df)
input_scaled
```

```python
# Predict the churn
prediction = model.predict(input_scaled)
```
